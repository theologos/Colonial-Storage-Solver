```python
import math
import pandas as pd

def data_topology(df_tickets):
    test = []
    df_1 = df_tickets
    for cycle in df_1.Cycle.unique():
        d = df_1[(df_1['Cycle'] == cycle) & (df_1['Type'] == 'Out')]
        for tank in d.Tank.unique():
            d2 = d[d["Tank"] == tank]
            for index, row in d2.iterrows():
                test.append([int(tank), row['Line'], row['Product']])

    my_dict = {}            
    for tank, line, prod in test:
        if tank not in my_dict:
            my_dict[tank] = [[line, prod]]
        else:
            if [line, prod] not in my_dict[tank]:
                my_dict[tank].append([line, prod])

    topology = my_dict
    return (topology)



def data_volume(df_tickets):

    ret = {}

    #--------------------------------------------------------------------------------------------
    #
    #
    df_1 = df_tickets
    df_1 = df_1.groupby(['Cycle', 'Line', 'Product', 'Type']).agg({'Volume': ['sum']}).reset_index()
    df_1.columns = [i[0]+"_"+i[1] for i in df_1.columns]
    df_1 = df_1[df_1['Product_'].isin(['A', 'D', '54', '62'])]
    df_1.columns = df_1.columns.str.rstrip('_')

    VolOut = {}
    VolIn  = {}
    for cycle in df_1.Cycle.unique():

        CycleVolOut2 = {}
        CycleVolIn2 = {}
        # CycleVolOut
        d = df_1[(df_1['Cycle'] == cycle) & (df_1['Type'] == 'Out')]
        for line in d.Line.unique():
            d2 = d[d["Line"] == line]
            temp = {}
            for index, row in d2.iterrows():
                temp[row['Product']] = math.ceil(row['Volume_sum'])
                CycleVolOut2[line] = temp

        # CycleVolIn
        d = df_1[(df_1['Cycle'] == cycle) & (df_1['Type'] == 'In')]
        for line in d.Line.unique():
            d2 = d[d["Line"] == line]
            temp = {}
            for index, row in d2.iterrows():
                temp[row['Product']] = math.ceil(row['Volume_sum'])
                CycleVolIn2[line] = temp

        VolOut[cycle] = CycleVolOut2
        VolIn[cycle]  = CycleVolIn2 
    ret['VolIn']  = VolIn
    ret['VolOut'] = VolOut

    #--------------------------------------------------------------------------------------------
    #
    #
    df_1 = df_tickets
    df_1['Rank'] = (df_1.groupby(['Cycle', 'Tank'])['Start']
                          .rank(method='dense', ascending=True)
                          .astype(int)
                 )

    df_1 = df_1.sort_values(by=['Cycle', 'Tank', 'Start'])
    df_1 = df_1[df_1.Rank == 1][['Cycle', 'Tank', 'Product', 'Volume_Tank']]
    VolExist = {}; 
    for cycle in df_tickets.Cycle.unique():
        VolExist_temp = {};
        d = df_1[df_1['Cycle'] == cycle]
        for index, row in d.iterrows():
            temp = {}; temp[row['Product']] = math.ceil(row['Volume_Tank'])
            VolExist_temp[row['Tank']] = temp
        VolExist[cycle] = VolExist_temp   
    ret['VolExist'] = VolExist   
    
    #--------------------------------------------------------------------------------------------
    #
    #
    VolLineFlows = {}
    for cycle_ in df_tickets.Cycle.unique():
    
        dataframes_to_append = []
        confs = df_tickets[df_tickets['Cycle'] == cycle_][['Cycle', 'Line', 'Product']]
        confs = confs.drop_duplicates()

        for index, row in confs.iterrows():

            cycle = row['Cycle']
            line  = row['Line']
            prod  = row['Product']

            df_temp = df_tickets[(df_tickets['Cycle'] == cycle) & (df_tickets['Line'] == line) & (df_tickets['Product'] == prod)]

            df_temp = df_temp.sort_values(by='Start Date/Time', ascending=True)
            df_temp = df_temp[['Start Date/Time', 'End Date/Time', 'Tank', 'Volume', 'Vol_per_Hr', 'Time_in_hours', 'Product', 'Line', 'Cycle']]
            df_temp['Start'] = df_temp['Start Date/Time'].dt.floor('H')
            df_temp['End']   = df_temp['End Date/Time'].dt.floor('H') - pd.Timedelta(hours=1)
            df_temp['Diff']  = ((df_temp['End'] - df_temp['Start']).dt.total_seconds() / 3600).astype(int) + 1

            for index, row in df_temp.iterrows():
                hourly_range = pd.date_range(start=row['Start'], end=row['End'], freq='H')
                a = pd.DataFrame({'Datetime': hourly_range})
                a['Volume']     = row['Volume']
                a['Diff']       = row['Diff']
                a['Cycle']      = row['Cycle']
                if row['Diff'] > 0:
                    a['Hourly_Vol'] = row['Volume'] / row['Diff']
                    a['Hourly_Vol'] = a['Hourly_Vol'].round(3)
                else:
                    a['Hourly_Vol'] = 0

                a['Product']    = row['Product']
                a['Line']       = row['Line']
                dataframes_to_append.append(a)

            result_df = pd.concat(dataframes_to_append, ignore_index=True)

            # Define a function to get the minimum value by group
            def get_min_by_group(group):
                return group.min()
            # Use transform to create a new column with minimum values by category
            result_df['Datetime_Min'] = result_df.groupby('Cycle')['Datetime'].transform(get_min_by_group)
            result_df['Time'] = (result_df['Datetime'] - result_df['Datetime_Min']).dt.total_seconds() / 3600
            result_df['Time'] = result_df['Time'].apply(round).astype(int)

            VolLineFlows[cycle_] = result_df   
   
    ret['VolLineFlows'] = VolLineFlows

    print('Done')
    return (ret)
```
