```python
import pandas as pd

# T         = 132
# Time      = list(range(T))

class DataLocation:
    def __init__(self, name):
        self.Name      = name
        
        self.get_tanks()
        #self.Tanks     = Tanks
        self.Tanks_073 = [tank for tank in self.Tanks if tank <= 339]
        self.Tanks_085 = [tank for tank in self.Tanks if tank > 339]
        #self.Capacity = Capacity
        #self.topo_i   = topo_i
        #self.topo_o   = topo_o
        
        self.Bounds   = {
            "01": {"l": 25,   "u": 28},
            "02": {"l": 6,    "u": 10},
            "13": {"l": 2,    "u": 2},
            "14": {"l": 2,    "u": 2.4}, 
            "15": {"l": 4,    "u": 5},
            "16": {"l": 2,    "u": 2.15},
            "17": {"l": 4,    "u": 5},
            "18": {"l": 7,    "u": 8.7},
            "19": {"l": 4,    "u": 5.1},
            "20": {"l": 1,    "u": 1.75},
            "1A": {"l": 1,    "u": 5},
            "2A": {"l": 1,    "u": 5}
           }
        
        
        self.topo_i = {
                  310: {'01': {'A': 0}},
                  311: {'01': {'A': 0}},
                  312: {'01': {'D': 0}},
                  313: {'01': {'A': 0}},
                  314: {'01': {'A': 0}},
                  316: {'01': {'D': 0}},
                  317: {'01': {'A': 0}},
                  330: {'01': {'A': 0}},
                  331: {'01': {'A': 0}},
                  332: {'01': {'A': 0}},
                  333: {'01': {'A': 0}},
                  334: {'01': {'A': 0}},
                  336: {'01': {'A': 0}},
                  337: {'01': {'D': 0}},
                  338: {'01': {'A': 0}},
                  339: {'01': {'A': 0}},
                  350: {'02': {'54': 0}},
                  351: {'02': {'54': 0}},
                  352: {'02': {'54': 0}},
                  353: {'02': {'54': 0}},
                  354: {'02': {'54': 0}},
                  360: {'02': {'62': 0}},
                  361: {'02': {'62': 0}},
                  363: {'02': {'62': 0}},
                  370: {'02': {'54': 0}},
                  371: {'02': {'62': 0}},
                  373: {'02': {'62': 0}},
                  374: {'02': {'62': 0}},
                  375: {'02': {'62': 0}},
                  376: {'02': {'62': 0}}
        }

        self.topo_o = {
                      310: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      311: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      312: {'13': {'D': 0}, '14': {'D': 0}, '15': {'D': 0}, '17': {'D': 0}, '18': {'D': 0}, '19': {'D': 0}, '1A': {'D': 0}},
                      313: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      314: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      316: {'13': {'D': 0}, '14': {'D': 0}, '15': {'D': 0}, '17': {'D': 0}, '18': {'D': 0}, '19': {'D': 0}, '1A': {'D': 0}},
                      317: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      330: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      331: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      332: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      333: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      334: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},  
                      336: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      337: {'13': {'D': 0}, '14': {'D': 0}, '15': {'D': 0}, '17': {'D': 0}, '18': {'D': 0}, '19': {'D': 0}, '1A': {'D': 0}},  
                      338: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      339: {'13': {'A': 0}, '14': {'A': 0}, '15': {'A': 0}, '17': {'A': 0}, '18': {'A': 0}, '19': {'A': 0}, '1A': {'A': 0}},
                      350: {'13': {'54': 0}, '14': {'54': 0}, '16': {'54': 0}, '17': {'54': 0}, '18': {'54': 0}, '19': {'54': 0}, '20': {'54': 0}, '2A': {'54': 0}},
                      351: {'13': {'54': 0}, '14': {'54': 0}, '16': {'54': 0}, '17': {'54': 0}, '18': {'54': 0}, '19': {'54': 0}, '20': {'54': 0}, '2A': {'54': 0}},
                      352: {'13': {'54': 0}, '14': {'54': 0}, '16': {'54': 0}, '17': {'54': 0}, '18': {'54': 0}, '19': {'54': 0}, '20': {'54': 0}, '2A': {'54': 0}},
                      353: {'13': {'54': 0}, '16': {'54': 0}, '20': {'54': 0}},
                      354: {'16': {'54': 0}, '18': {'54': 0}, '20': {'54': 0}},
                      360: {'13': {'62': 0}, '14': {'62': 0}, '17': {'62': 0}, '18': {'62': 0}, '19': {'62': 0}, '20': {'62': 0}, '2A': {'62': 0}},
                      361: {'13': {'62': 0}, '14': {'62': 0}, '17': {'62': 0}, '18': {'62': 0}, '19': {'62': 0}, '20': {'62': 0}, '2A': {'62': 0}},
                      363: {'13': {'62': 0}, '14': {'62': 0}, '17': {'62': 0}, '18': {'62': 0}, '19': {'62': 0}, '20': {'62': 0}, '2A': {'62': 0}},
                      370: {'13': {'54': 0}, '14': {'54': 0}, '17': {'54': 0}, '18': {'54': 0}, '19': {'54': 0}, '20': {'54': 0}, '2A': {'54': 0}},
                      371: {'13': {'62': 0}, '14': {'62': 0}, '17': {'62': 0}, '18': {'62': 0}, '19': {'62': 0}, '20': {'62': 0}, '2A': {'62': 0}},
                      373: {'13': {'62': 0}, '14': {'62': 0}, '17': {'62': 0}, '18': {'62': 0}, '19': {'62': 0}, '20': {'62': 0}, '2A': {'62': 0}},          
                      374: {'13': {'62': 0}, '14': {'62': 0}, '17': {'62': 0}, '18': {'62': 0}, '19': {'62': 0}, '20': {'62': 0}, '2A': {'62': 0}},
                      375: {'13': {'62': 0}, '14': {'62': 0}, '17': {'62': 0}, '18': {'62': 0}, '19': {'62': 0}, '20': {'62': 0}, '2A': {'62': 0}},
                      376: {'13': {'62': 0}, '14': {'62': 0}, '17': {'62': 0}, '18': {'62': 0}, '19': {'62': 0}, '20': {'62': 0}, '2A': {'62': 0}}
        }
        
        self.validate_topologies(self.topo_i, self.topo_o)
        self.validate_tanks(self.topo_i, self.Tanks)
        self.validate_lines(self.topo_i, self.topo_o, self.Bounds)
        
    def get_tanks(self):
        
        filename = 'data_new_1/capacities_ATJ.csv'
        self.Capacity = {}
        df1 = pd.read_csv (filename, encoding = "ISO-8859-1")
        for index, row in df1.iterrows():
            self.Capacity[int(row['Tank'])] = round(int(1000 * row['Working'])/1000)

        for tank in self.Capacity:
            self.Capacity[tank] = self.Capacity[tank] + 10

        self.Tanks = list(self.Capacity.keys())
        rem = [315, 335, 372]
        self.Tanks = list(set(self.Tanks) - set(rem))
    
    
    def validate_topologies(self, topo_i, topo_o):
        if not list(topo_i.keys()) == list(topo_o.keys()):
            raise ValueError("Error: Input and Output topology keys are not the same.")

    def validate_tanks(self, topo_i, Tanks):
        if not list(topo_i.keys()) == sorted(Tanks):
            raise ValueError("Error: Topology and Tank keys do not match.")
            
    def validate_lines(self, topo_i, topo_o, Bounds):
        keys_o = set()
        for k, v in topo_o.items():
            keys_o.update(v.keys())

        keys_i = set()
        for k, v in topo_i.items():
            keys_i.update(v.keys()) 
            
        keys = list(keys_i) + list(keys_o)   
        if not sorted(keys) == sorted(list(Bounds.keys())):
            raise ValueError("Error: There is an issue with Lines.")
```
