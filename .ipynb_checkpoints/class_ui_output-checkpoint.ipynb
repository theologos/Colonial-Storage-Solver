```python
# Import necessary libraries
import pandas as pd
import plotly.express as px

class Scatterplot:
    def __init__(self, id_):
        self.id = id_
                
    def render(self):
        rt = html.Div([
                dcc.Graph(id='graph-scatterplot')
            ])
            
        @app.callback(
            Output('graph-scatterplot', 'figure'),
            [Input('dropdown-line-schedule', 'value')]
        )
        def update_graph(schedule):

            # Load the data            
            df = data_optimizations.getOptimization("ATJ", schedule).CycleVolLineFlows
            
            # Convert 'Datetime' column to datetime type
            df['Datetime'] = pd.to_datetime(df['Datetime'])

            # Create the timeline graph using Plotly
            fig = px.scatter(df, x='Datetime', y='Line', size='Hourly_Vol', 
                             color='Product', hover_data=['Hourly_Vol', 'Product'],
                             labels={'Line': 'Line', 'Datetime': 'Time'},
                             title='Hourly Volume'
                            ) 
            
            return fig
        
        return rt
```
